```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('./dataset/malaria_detection.csv') 
df.head(10)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
df["label"]=enc.fit_transform(df["label"])
df.head(10)
```

```python
from skimage.io import imread,imshow
from skimage.transform import resize
```

```python
import os
base_path=os.path.join('./dataset/cell_images/')
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y=train_test_split(df["filename"], df["label"], test_size=0.2, random_state=42)
```

```python
train_x.size, test_x.size
```

```python
img_x=[]
img_y=[]
```

```python
for i in range(len(train_x)):
    img=imread(os.path.join(base_path,train_x.iloc[i]))
    img_x.append(img.shape[0])
    img_y.append(img.shape[1])
avg_x=sum(img_x)/len(img_x)
avg_y=sum(img_y)/len(img_y)
avg_x,avg_y
```

```python
img_dim=(132,132)
```

```python
from skimage.feature import hog
from skimage.transform import resize
```

```python
train_features = []
for i in train_x:
    image = imread(os.path.join(base_path,i))
    image = resize(image,img_dim)
    features, hog_image = hog(image, orientations=9, pixels_per_cell=(8, 8),cells_per_block=(2, 2), visualize=True, multichannel=True)
    train_features.append(image)
```

```python
test_features = []
for i in test_x:
    image = imread(os.path.join(base_path,i))
    image = resize(image,img_dim)
    features, hog_image = hog(image, orientations=9, pixels_per_cell=(8, 8),cells_per_block=(2, 2), visualize=True, multichannel=True)
    test_features.append(image)
```

```python
train_features=np.array(train_features)
test_features=np.array(test_features)
```

```python
np.save('train_features', train_features)
np.save('test_features', test_features)
```

```python
images=[]
for i in train_x:
    images.append(imread(os.path.join(base_path,i)))
images
```

```python
images=np.array(images)
images.shape
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, Flatten, MaxPooling2D, Dense, Dropout
from keras.layers.advanced_activations import LeakyReLU
```

```python
model=Sequential()
model.add(Conv2D(32,kernel_size=(3,3), input_shape=(130,130,3), activation='relu', padding='same'))
model.add(LeakyReLU(alpha=0.1))
model.add(MaxPooling2D((2,2), padding='same'))
model.add(Flatten())
model.add(Dense(128, activation='linear'))
model.add(LeakyReLU(alpha=0.1))
model.add(Dense(10, activation='softmax'))
```

```python
train_x2=train_x[1:]
train_y2=train_y[1:]
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
hist=model.fit(images, train_y2, epochs=3, batch_size=64,validation_data=(test_x, test_x))
```
